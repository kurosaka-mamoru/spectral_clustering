```lua
torch.setdefaulttensortype('torch.FloatTensor')
```

```lua
x = torch.rand(100, 100)
k = 10
eps = 1e-6
```

```lua
D = torch.diag(torch.sum(x, 1):squeeze())
```

```lua
D_safe = torch.diag(D)
divided = torch.Tensor(D_safe:size()):fill(eps + 1.0)
D_coef = torch.diag(torch.cdiv(divided, D_safe))
```

```lua
D_coef
```

```lua
L = D - x
```

```lua
L
```

```lua
L = D_coef * L
```

```lua
L
```

```lua
torch.eig(L, D)
```
